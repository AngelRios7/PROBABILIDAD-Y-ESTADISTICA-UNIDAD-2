```python
import math as mt
from itertools import permutations, combinations
def calcular_conteo_completo(n, r):
    if r > n:
        return "Error: r no puede ser mayor que n"

    # 1. Cálculo de Permutaciones
    # Fórmula: n! / (n - r)!
    p_n_r = mt.factorial(n) / mt.factorial(n - r)
    # 2. Cálculo de Combinaciones
    # Fórmula: n! / (r! * (n - r)!)
    c_n_r = mt.factorial(n) / (mt.factorial(r) * mt.factorial(n - r))
    print(f"Análisis para n={n} y r={r}")
    print(f"Permutaciones (P): {int(p_n_r)}")
    print(f"Combinaciones (C): {int(c_n_r)}")
    print("-" * 35)
    return p_n_r, c_n_r

# Resolucion del Ejercicio de los asistentes
# "Diez asistentes disponibles (n=10) para 4 preguntas (r=4)"
n_asistentes = 10
r_preguntas = 4

# Lista de identificadores para los asistentes
asistentes = [f"Asistente {i}" for i in range(1, n_asistentes + 1)]

# Ejecutar calculos
p, c = calcular_conteo_completo(n_asistentes, r_preguntas)

print(f"Como cada pregunta requiere un asistente diferente, el orden importa.")
print(f"Existen {int(p)} formas de asignarlos.")

```
